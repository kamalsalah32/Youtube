**Lists**

```python
l = ["orange", "mango", "kiwi", "pineapple", "banana"]
# Ordered
# Changeable
# Allow Duplicates


# Get item
print(l[1])

# Update Item
l[0] = "apple"
print(l)
print("----------")

# Add Item 
# - append(item)
l.append("orange")
print(l)

# - insert(index, item)
l.insert(1, "avocado")
print(l)
l = l + ["pear"]
print(l)
print("----------")

# Remove Item
# - remove(item)
l.remove("pear")
print(l)

# - pop(index) / pop()
item = l.pop(0)
print(l)
print(item)
item1 = l.pop()
print(l)
print(item1)

# - del
del l[2]
print(l)

# - clear()
#l.clear()
#print(l)
print("----------")

# Loop over list
# - for _ in _
for x in l:
  print(x)

print("---")
# - for _ in range(_)
for i in range(0, len(l)):
  print(l[i])

print("---")
# - while _
i = 0
while i < len(l):
  print(l[i])
  i += 1

print("---")
# - list comprehension [ _ for _ in _ ]
[print(x) for x in l]

print("---")
# List Comprehension: newlist = [expression for item in iterable if condition == True]
newList = [x for x in l if 'n' in x]
print(newList)
print("----------")

# sort()
list1 = [100, 50, 65, 82, 23]
print("List1: ", list1)
list1.sort()
print("List1: ", list1)

# sort(reverse = True)
list1.sort(reverse = True)
print("List1: ", list1)
print("----------")

# Copy list
# - copy()
'''
list2 = list1
print(list1) 
print(list2)
print("---")
list2[0] = 200
print(list2)
print(list1)
'''
list2 = list1.copy()
print("List1: ", list1) 
print("List2: ", list2)
print("---")
list2[0] = 200
print("List1: ", list1) 
print("List2: ", list2)
print("----------")

# - constructor
list3 = list(list1)
print("List1: ", list1) 
print("List3: ", list3)
print("---")
list3[0] = 200
print("List1: ", list1) 
print("List3: ", list3)

print("----------")

# Join two lists
# - '+'
list4 = list1 + list3
print("List4: ", list4)

# - append(item)
for x in list3:
  list1.append(x)
print("List1: ", list1) 

# - extend(list)
list2.extend(list3)
print("List2: ", list2)
```

**Tuples**

```python
t1 = ("orange", "mango", "kiwi", "pineapple", "banana")
t2 = (1, 2, 3)
# Unchangeable
# Ordered

# Get Item
print(t1[1])

# Update Item
print("t1: ", t1)
l1 = list(t1)
print("l1: ", l1)
l1[0] = "apple"
t1 = tuple(l1)
print("t1: ", t1)
print("----------")

# Add Item
l2 = list(t2)
l2.append(5)
l2 += [6]
l2.insert(0, 100)
print("t2: ", l2)
t2 = tuple(l2)
print("t2: ", t2)
print("----------")

# Join tuples
t3 = t1 + t2
print("t3: ", t3)
print("----------")

# Unpack tuples
(x, y, z, w, v) = t1
print(x)
print(y)
print(z)
print(x)
print(v)
print("----------")

# Loop over tuples
for x in t1:
  print(x)

print("---")
for i in range(len(t1)):
  print(t1[i])

print("---")
[print(x) for x in t1]  

print("----------")

# Methods for tuples and lists
# - count(item)
print(t1.count("apple"))
# - index(item)
print(t1.index("pineapple"))
```

**Set**

```python
s1 = {"orange", "mango", "kiwi", "pineapple", "banana"}
s2 = {1, 2, 3}
# Unordered
# No Duplicates

# Add Item
s1.add("apple")
print("s1: ", s1)
s1.add("apple")
print("s1: ", s1)
print("----------")

# Remove Item
# - remove()
s1.remove("kiwi")
print("s1: ", s1)
print("---")

# - discard()
s1.discard("banana")
print("s1: ", s1)
print("---")

# - pop()
s1.pop()
print("s1: ", s1)
print("----------")

# Check if item exists
print("kiwi" in s1)
print("----------")

# Loop over set

for x in s1:
  print(x)

print("---")
[print(x) for x in s1]

print("----------")

# Union
s3 = s1.union(s2)
print("s3: ", s3)
print("----------")

# Intersection
# - intersection()

s2.add("orange")
print("s1: ", s1)
print("s2: ", s2)
s4 = s1.intersection(s2)
print("s4: ", s4)

# - intersection_update()
#s1.intersection_update(s2)
#print(s1)
print("----------")


# Symmetric Difference
# - symmetric_difference()
s5 = s1.symmetric_difference(s2)
print("s5: ", s5)
print("---")
# - symmetric_difference_update()
s1.symmetric_difference_update(s2)
print("s1: ", s1)
print("----------")

# issubset()
s_new = {1, 2}
print("s_new: ", s_new)
print("s1: ", s1)
print(s_new.issubset(s1))
print(s1.issubset(s_new))
print("----------")

# issuperset()
print(s_new.issuperset(s1))
print(s1.issuperset(s_new))
print("----------")

# isdisjoint()
print(s1.isdisjoint(s_new))
```
